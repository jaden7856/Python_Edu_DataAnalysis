```python
# 01) Dataset
baseball_player =  pd.DataFrame(
    {
        "player_Name":
        ["권혁","이범호","필","박병호","양준혁"],
        "team_No":
        [1993,2001,2001,2010,None]
    }
)
baseball_team = pd.DataFrame(
    {
        "team_Name":["한화","두산","기아","넥센"],
        "team_Home":["대전","서울","광주","고척"],
        "team_Id":[1993,1999,2001,2010]
    }
)
```

```python
baseball_player
```

```python
baseball_team
```

```python
# Left Outer join
pd.merge(baseball_player,baseball_team,how="left",left_on="team_No",right_on="team_Id")
```

```python
# Right Outer join
pd.merge(baseball_player,baseball_team,how="right",left_on="team_No",right_on="team_Id")
```

```python
#Inner join
pd.merge(baseball_player,baseball_team,how="inner",left_on="team_No",right_on="team_Id")
```

```python
# left excluding join
# 01) left join
temp = pd.merge(baseball_player,baseball_team,how="left",left_on="team_No",right_on="team_Id")
# 02) select NaN
temp[temp["team_Id"].isnull()]
```

```python
# right excluding join
# 01) right join
temp2 = pd.merge(baseball_player,baseball_team,how="right",left_on="team_No",right_on="team_Id")
# 02) select NaN
temp2[temp2["player_Name"].isnull()]
```

```python
pd.merge(baseball_player,baseball_team,how="outer",left_on="team_No",right_on="team_Id")
```

```python
# full outer join
temp3 = pd.merge(baseball_player,baseball_team,how="outer",left_on="team_No",right_on="team_Id")

temp3[temp3["player_Name"].isnull() | temp3["team_Id"].isnull()]
```

```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_excel("C:\\Users\\1-11\\Jupyter\\sampledata\\11_sales-funnel.xlsx")
data
```

```python
data["Status"]=data["Status"].astype("category")

data["Status"].cat.set_categories(["won","pending","presented","declined"],inplace=True)

data.head()
```

```python
# Name을 가로줄을 기준으로 하고 데이터를 보자 -> 숫자인 컬럼들만 들어온다
pd.pivot_table(data,index=["Name"]) # index에 여러 개가 들어올 수 있으니까 리스트로 받는다.
```

```python
# 가로줄 여려 계층 구조로 사용할 경우
pd.pivot_table(data, index=["Name","Rep","Manager"])
```

```python
# 2개만 가로줄 선택
pd.pivot_table(data, index=["Manager","Rep"])
```

```python
pd.pivot_table(data, index=["Manager","Rep"],values=["Price"])
```

```python
# pivot with aggfunc : sum
import numpy as np
pd.pivot_table(data,index=["Manager","Rep"], values=["Price"], aggfunc=np.sum)
```

```python
# pivot with aggfunc --> make new column
pd.pivot_table(data, index=["Manager","Rep"], values=["Price"], aggfunc=[np.mean,len])
```

```python
# 세로줄에 대한 계층
pd.pivot_table(data, index=["Manager","Rep"], values=["Price"],columns=["Product"], aggfunc=[np.sum])
```

```python
pd.pivot_table(data, index=["Manager","Rep"], values=["Price"],columns=["Product"], aggfunc=[np.sum], fill_value=0)
```

```python
# 여러 칼럼에 여러 계층 구조
pd.pivot_table(data, index=["Manager","Rep"], values=["Price","Quantity"],columns=["Product"], aggfunc=[np.sum], fill_value=0)
```

```python
# 품목에 대한 정보를 가로 인덱스로
pd.pivot_table(data, index=["Manager","Rep","Product"], values=["Price","Quantity"], aggfunc=[np.sum], fill_value=0)
```

```python
# Total option
pd.pivot_table(data, index=["Manager","Rep","Product"], values=["Price","Quantity"],
               aggfunc=[np.sum,np.mean], fill_value=0, margins=True)
```

```python
# row : manager / status
# column : Quantity [len] / price [sum]
# column category : product
pd.pivot_table(data, index=["Manager","Status"], columns=["Product"], values=["Price","Quantity"],
               aggfunc={"Quantity":len,"Price":np.sum}, fill_value=0)
```

```python
# aggfunc 에서도 계층 구조 적용
pd.pivot_table(data, index=["Manager","Status"], columns=["Product"], values=["Quantity","Price"],
              aggfunc={"Quantity":len,"Price":[np.sum, np.mean]}, fill_value=0)
```

```python
table = pd.pivot_table(data, index=["Manager","Status"], columns=["Product"], values=["Quantity","Price"],
              aggfunc={"Quantity":len,"Price":[np.sum, np.mean]}, fill_value=0)

table.query('Manager == ["Debra Henley"]')
```

```python
pd.crosstab([data["Manager"], data["Rep"]], [data["Price"], data["Quantity"]])
```
